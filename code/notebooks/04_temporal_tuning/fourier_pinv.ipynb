```python
import dlop_ldn_function_bases as bases
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = bases.mk_fourier_basis(21, 2000)
H_red = H[:, 1000:]
```

```python
Hp = np.linalg.pinv(H_red, rcond=1e-6)
```

```python
fig, ax = plt.subplots()
ax.imshow(Hp @ H_red)
ax.set_aspect('auto')
```

```python
enc = H_red[:, 0]
dec = Hp[0]
```

```python
fig, ax = plt.subplots()
ax.plot((dec @ H_red))
```

```python
fig, ax = plt.subplots()
ax.imshow((dec @ H).reshape(1, -1), interpolation='None')
ax.set_aspect('auto')
```

```python
fig, ax = plt.subplots()
ax.imshow(np.outer(H[:, 500], Hp[500]), interpolation='None')
ax.set_aspect('auto')
```
