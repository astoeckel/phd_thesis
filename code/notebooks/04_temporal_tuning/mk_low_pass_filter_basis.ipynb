```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mk_low_pass_filter_basis(q, N, tau0=2e-3, tau1=100.0):
    ts = np.linspace(0, 1, N)
    taus = np.geomspace(tau0, tau1, q)
    res = np.array([np.exp(-ts / tau) for tau in taus])
    res /= np.sqrt(np.array([np.inner(res[i], res[i]) for i in range(q)])[:, None])
    return res
```

```python
B = mk_low_pass_filter_basis(16, 100)
fig, ax = plt.subplots()
ax.plot(B.T);
```
