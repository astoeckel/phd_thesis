```python
import nengo
import nengo.utils.ensemble
import numpy as np
import matplotlib.pyplot as plt
```

```python
import h5py
with h5py.File("low_firing_rates_dynamics_sweep.h5", "r") as f:  
    data = f["results"][()]
    rates = f["rates"]
    fs = f["fs"]

    N_types = data.shape[1]
    fig, axs = plt.subplots(1, N_types, figsize=(N_types * 5.0, 5.0))
    for i in range(N_types):
        avg = 10 * np.mean(np.log10(np.abs(data[:, i])), axis=0)
        #avg = 10 * np.log10(np.mean(np.abs(data[:, i]), axis=0))
        n_freq = avg.shape[1]
        axs[i].imshow(avg[::-1, (n_freq // 2):].T, vmin=-5, vmax=0.3, extent=[
            100, 10,
            0, 100
        ], origin='lower', cmap='inferno')
        axs[i].set_aspect("auto")
        axs[i].set_xticks([100, 10])
        axs[i].set_xticks(np.power(10, np.linspace(1, 2, 11)), minor=True)
        axs[i].set_ylabel('Frequency $f$ (Hz)')
        axs[i].set_xlabel('Maximum firing rate $a_\\mathrm{max}$ ($\\mathrm{s}^{-1}$)')
        #axs[i].set_xticklabels(np.logspace(2, 1, 5))
```
