```python
import numpy as np

import sys, os
import h5py

import matplotlib.pyplot as plt
```

```python
with h5py.File(os.path.join('..', '..', 'data', 'nonnegative_experiment_large2.h5'), 'r') as f:
    p_excs = f['p_excs'][()]
    errs = f['errs'][()]
```

```python
errs.shape
```

```python
fig, ax = plt.subplots()
Ei = 2
Fi = 0
ax.plot(p_excs, np.mean(errs[Ei, :, 1, 0, Fi], axis=-1), label='Dale, intr. bias')
ax.plot(p_excs, np.mean(errs[Ei, :, 1, 1, Fi], axis=-1), label='Dale, dec. bias')
ax.axhline(np.mean(errs[Ei, 0, 0, 0, Fi], axis=-1), label='No dale, intr. bias')
ax.axhline(np.mean(errs[Ei, 0, 0, 1, Fi], axis=-1), label='No dale, dec. bias')
ax.set_yscale('log')
#ax.set_ylim(5e-3, 2e0)
ax.set_ylim(5e-1, 1e2)
ax.legend()
```
