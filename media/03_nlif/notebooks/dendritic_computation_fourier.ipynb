```python
def coeffs(d, mu=1, sigma=1.0, rng=np.random):
    ds = np.arange(0, d)
    mask = np.exp(-np.square(ds - mu) / np.square(sigma + 1e-3))
    mask2d = np.outer(mask, mask)
    mask2d /= np.sum(mask2d)
    return mask
```

```python
fig, ax = plt.subplots()
for sigma in np.logspace(-0.4, 2, 10):
    ax.plot(coeffs(16, 0, sigma))
ax.set_ylim(0, 1)
```
